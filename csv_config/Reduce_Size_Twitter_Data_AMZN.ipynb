```python
tweets_only_path = '/content/drive/MyDrive/Uni stuff/Text Mining/Tweet.csv'
company_tweet_path = '/content/drive/MyDrive/Uni stuff/Text Mining/Company_Tweet.csv'
company_path = '/content/drive/MyDrive/Uni stuff/Text Mining/Company.csv'

df_tweets_only = pd.read_csv(tweets_only_path)
df_company_tweet = pd.read_csv(company_tweet_path)
df_company = pd.read_csv(company_path)
```

```python
df_company_tweet["ticker_symbol"].unique()
```

```python
values_to_filter = ['TSLA', 'GOOG', 'GOOGL', 'MSFT', 'APPL']
df_company_tweet_filtered = df_company_tweet[~df_company_tweet["ticker_symbol"].isin(values_to_filter)]
```

```python
df_company_name_tweet_filtered = pd.merge(df_company, df_company_tweet_filtered, on='ticker_symbol', how='inner')
df_company_name_tweet_column_del = df_company_name_tweet_filtered.drop('ticker_symbol', axis=1)
df_tweets_combined = pd.merge(df_tweets_only, df_company_name_tweet_column_del, on='tweet_id', how='inner')
df_tweets = df_tweets_combined.drop('tweet_id', axis=1)
```

Convert twitter time to day-month-year representation

```python
df_tweets['date'] = pd.to_datetime(df_tweets['post_date'], unit='s').dt.floor('d')
```

```python
df_tweets_2015 = df_tweets[df_tweets['date'].dt.year < 2016]
```

```python
print(df_tweets_2015.head())
file_path = '/content/drive/MyDrive/Uni stuff/Text Mining/smaller_Tweets_AMZN.csv'
df_tweets_2015.to_csv(file_path, index=False)
print(len(df_tweets_2015))
```
